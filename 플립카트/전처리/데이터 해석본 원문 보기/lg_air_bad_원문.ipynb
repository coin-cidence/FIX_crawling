```python
import pandas as pd
```

```python
df_LG_airconditioner = pd.read_csv('df_LG_airconditioner.csv')

# Filter rows where 'star' is 1 or 2
filtered_df_LG_airconditioner = df_LG_airconditioner[df_LG_airconditioner['star'].isin([1, 2])]

# Define the keywords to search for in the 'text' column
keywords = ['cooling']

# Filter rows where 'text' contains any of the keywords
filtered_df_LG_airconditioner = filtered_df_LG_airconditioner[filtered_df_LG_airconditioner['text'].str.contains('|'.join(keywords), case=False, na=False)]

# Display the filtered rows
filtered_df_LG_airconditioner.reset_index(drop=True, inplace=True)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
filtered_df_LG_airconditioner
```
