```python
import pandas as pd

def assign_state(city_name):
    # Check if city_name is NaN
    if pd.isna(city_name):
        # Loop through the dictionary to find the state for NaN values
        for state, cities in india_districts.items():
            if city_name in cities:
                return state
        return None  # Return None if city_name is not found in any state's cities
    else:
        # Return the original city_name if it's not NaN
        return city_name


df['state'] = df['city'].apply(assign_state)
```

```python
import pandas as pd

def assign_state(city_name):
    # Check if city_name is NaN
    if pd.isna(city_name):
        # Loop through the dictionary to find the state for NaN values
        for state, cities in town.items():
            if city_name in cities:
                return state
        return None  # Return None if city_name is not found in any state's cities
    else:
        # Return the original city_name if it's not NaN
        return city_name


df['state'] = df['city'].apply(assign_state)
```
