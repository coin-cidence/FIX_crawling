```python
import pandas as pd
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords
# 워드클라우드 import
from wordcloud import WordCloud
# 데이터를 시각화 해주는 라이브러리
import matplotlib.pyplot as plt
```

```python
df_lg_air_bad = pd.read_csv('df_lg_air_bad.csv')
df_lg_air_bad_top20 = df_lg_air_bad.nlargest(20, 'freq')

# Step 3: Example list of words to remove
words_to_remove = ['ac', 'product', 'buy', 'lg','worst','1','bad','flipkart','good','even','unit','ton']

# Filter out the selected words from the top 20 list
df_lg_air_bad_top20_filtered = df_lg_air_bad_top20[~df_lg_air_bad_top20['word'].isin(words_to_remove)]


# Pie chart visualization
plt.figure(figsize=(10, 7))
plt.pie(df_lg_air_bad_top20_filtered['freq'], labels=df_lg_air_bad_top20_filtered['word'], autopct='%1.1f%%', startangle=140)
plt.title('df_lg_air_bad_top20_filtered')
plt.show()
```
