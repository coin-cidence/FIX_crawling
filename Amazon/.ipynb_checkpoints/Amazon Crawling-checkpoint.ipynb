```python
driver=webdriver.Chrome()
driver.get("https://www.amazon.in/s?k=lg&i=electronics&rh=n%3A976419031%2Cn%3A1389375031%2Cp_123%3A46658&s=exact-aware-popularity-rank&dc&language=en_IN&crid=2QA43FVMBHUNH&qid=1724642058&rnid=976419031&sprefix=l%2Caps%2C291&ref=sr_st_exact-aware-popularity-rank&ds=v1%3AgRnTmIKFL85z6OoGBaubSRC5FU%2FY2sKRTgFHNhuK%2Fy8")

button = driver.find_element(By.CLASS_NAME,"a-icon.a-icon-popover")
button.click()
```

```python
# ÌéòÏù¥ÏßÄ ÎÑòÍ∏∞Î©¥ÏÑú Î¶¨Ïä§Ìä∏ ÎΩëÍ∏∞
title_list=[]
content_list=[]
# Ìïú ÌéòÏù¥ÏßÄ ÏïàÏùò ÏùåÏãùÏ†ê Î¶¨Ïä§Ìä∏
while True:
    sum=0
    for i in range(15):
        try:
            #ÏùåÏãùÏ†êÎ™ÖÍ≥º Î≥ÑÏ†ê
            title=driver.find_elements(By.CLASS_NAME,'a-size-base.a-link-normal.review-title.a-color-base.review-title-content.a-text-bold')
            for i in title:
                title_list.append(i.text)
                   
            content=driver.find_elements(By.CLASS_NAME,'a-size-base.review-text.review-text-content')
            for i in content:
                content_list.append(i.text)
                

            

            try:
                button = driver.find_element(By.CLASS_NAME, "a-last")
                button.click()
                time.sleep(2)
                
            except NoSuchElementException:
                sum+=1
                break  # Î∞òÎ≥µÎ¨∏ÏùÑ ÏÇ¨Ïö©Ìï† Í≤ΩÏö∞, Ïù¥ Íµ¨Î¨∏ÏúºÎ°ú Î£®ÌîÑÎ•º Ï¢ÖÎ£åÌï©ÎãàÎã§.

                
        except:
            sum+=1
            break
            
    if sum==1:
        break
```

```python
# NLTK ÏòÅÏñ¥ Î∂àÏö©Ïñ¥ ÏÇ¨Ï†Ñ ÌôúÏö©, Î∂àÏö©Ïñ¥ Ï†úÍ±∞ÌïòÍ≥† Î¨∏Ïû•Ïùò Îã®Ïñ¥ ÎπàÎèÑÏàò Ï∂îÏ∂úÌïòÎäî Ìï®Ïàò 
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords

# NLTKÏóêÏÑú Î∂àÏö©Ïñ¥ Î™©Î°ù Îã§Ïö¥Î°úÎìú
nltk.download('stopwords')

def get_word_frequencies(texts, stop_words):
    words = []
    for text in texts:
        # Î™®Îì† Îã®Ïñ¥Î•º ÏÜåÎ¨∏ÏûêÎ°ú Î≥ÄÌôòÌïòÍ≥†, Îã®Ïñ¥Îßå Ï∂îÏ∂ú
        words.extend(re.findall(r'\b\w+\b', text.lower()))
    
    # Î∂àÏö©Ïñ¥Î•º Ï†úÏô∏Ìïú Îã®Ïñ¥Îßå ÎÇ®Í∏∞Í∏∞
    filtered_words = [word for word in words if word not in stop_words]
    
    # Îã®Ïñ¥ ÎπàÎèÑ Í≥ÑÏÇ∞
    word_counts = Counter(filtered_words)
    
    # ÎπàÎèÑÏàò ÏàúÏúºÎ°ú Ï†ïÎ†¨
    sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_word_counts

# ÏòÅÏñ¥ Î∂àÏö©Ïñ¥ Î™©Î°ù Í∞ÄÏ†∏Ïò§Í∏∞
stop_words = set(stopwords.words('english'))
```

```python
word_frequencies = get_word_frequencies(title_list, stop_words)

# Í≤∞Í≥º Ï∂úÎ†•
for word, freq in word_frequencies:
    print(f"{word}: {freq}")
```
